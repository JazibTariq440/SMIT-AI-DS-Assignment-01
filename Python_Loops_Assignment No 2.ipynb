# **Q8:** Ek list `numbers = [12, 45, 7, 23, 56, 90, 11]` mein se sabse bada (Maximum) number find karein loop ke zariye.

````python
# type your code here
numbers = [12, 45, 7, 23, 56, 90, 11]

max_num = numbers[0]

for num in numbers:
    if num > max_num:
        max_num = num

print ('Maximum number is', max_num)
````

# **Q12:** User se ek number lein aur check karein ke wo **Prime Number** hai ya nahi (using loops).

````python
# type your code here
number = int(input("enter num"))

is_prime = True

if number <= 1:
    is_prime = False
else:
    for i in range (2, number):
        if number % i == 0:
            is_prime = False
            break

if is_prime:
    print(number, "Prime Number")
else:
    print (number, "Not Prime Number")
````

# **Q16:** User se ek number input lein aur us ka **Factorial** find karein (e.g., 5! = 5*4*3*2*1).

````python
# type your code here
num = int(input("enter number"))

factorial = 1

for i in range (1, num + 1):
    factorial = factorial * i

print ("factorial =", factorial)
````

# **Q18:** Ek manual counting karein ke ek string (e.g., 'Learning Python is fun') mein kitne **Vowels** (a, e, i, o, u) hain.

````python
# type your code here
text = input("string")

vowels = 'aeiou'
count = 0

for char in text:
    if char.lower() in vowels:
        count = count + 1

print ("Total Vowels is", count)

    
````

# **Q20:** Ek 'Number Guessing Game' banayein jahan program ek number (1-20) fix kare aur user ko 3 chances de guess karne ke liye. `while loop` aur `break` ka use karein.

````python
# type your code here
secret = 14   # fixed number (1-20)

chances = 3

while chances > 0:
    guess = int(input("Guess the number (1-20): "))

    if guess == secret:
        print("Correct! Tum jeet gaye 🎉")
        break
    else:
        print("Galat guess ❌")

    chances = chances - 1

if chances == 0:
    print("Game Over 😢 Sahi number tha:", secret)
````
